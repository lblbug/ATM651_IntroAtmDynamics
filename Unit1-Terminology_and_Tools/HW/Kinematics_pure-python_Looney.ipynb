### Background
**Kinematics** is not dynamics, it is just a *description* of how 2D vector fields vary in 2D. 

* Assume LINEAR dependence of u,v on x,y 
    * This involves 4 coefficients $c_1, c_2, c_3, c_4$
* Also, add a constant (not variying with x,y) vector field U,V 

$$ u = U + c_1 x  +c_2 y$$
$$ v = V + c_3 x  +c_4 y$$

## Exercise 1 Answer:
* $ div = \frac{\partial u}{\partial x} + \frac{\partial v}{\partial y}= (U + c_1 x + c_2 y)_x + (V + c_3 x  +c_4 y)_y = c_1 + c_4$
* $ \zeta_{rel} = \frac{\partial v}{\partial x} - \frac{\partial u}{\partial y} =(V + c_3 x  +c_4 y)_x - (U + c_1 x  +c_2 y)_y = c_3 - c_2 $
* $ def1 = \frac{\partial u}{\partial x} - \frac{\partial v}{\partial y} = (U + c_1 x  +c_2 y)_x - (V + c_3 x  +c_4 y)_y = c_1 - c_4$
* $ def2 = \frac{\partial u}{\partial y} + \frac{\partial v}{\partial x} = (U + c_1 x  +c_2 y)_y + (V + c_3 x  +c_4 y)_x = c_2 + c_3$

```python
import numpy as np
import matplotlib.pyplot as plt

# Set up a coordinate grid centered on 0,0
n = 10
x,y = np.mgrid[-n:n+1, -n:n+1]
```

```python
# Pure divergence of unit magnitude
udiv,vdiv = 1*x, 1*y

# Pure vorticity of unit magnitude
uvor,vvor = -1*y, 1*x

# Pure def1 of unit magnitude
udef1,vdef1 = 1*x, -1*y

# Pure def2 of unit magnitude
udef2,vdef2 = 1*y, 1*x

# Pure U and V
uU, vU = 1, 0
uV, vV = 0, 1
```

## Answer to Exercise 3: 

Vorticity plus deformation (two curved flow patterns) makes a flow we could describe as...

When you have vorticity and deformation, there is a clear "asymptote" where the resulting wind vectors approaches 0. 

In the case of Deformation 1, this "asymptote" is on the x = y line. As you move further from the line, these wind vectors become opposite and stronger, yet close the the line, there is curvature, or vorticity, of the wind.On this line of x = y, there is no v component to the wind. Negative of the line (i.e. up and left), the v component is negative and the u component becomes more negative. Positive of the line (i.e. down and right), the v component is postivie, as is the u component. The magnitude of subsequent wind vectors increase the further from the line. 

In the case of Deformation 2, this "asymptote" is on the x = 0 line. The further from the line, the closer the 0 the u component is. On this line of x = 0, there is no v component. Negative of the line (i.e. left), the v component is negative. Positive of the line (i.e. right), the v component is postivie. The magnitude of subsequent wind vectors increase the further from the center.

## Answer to Exercise 4: 

Vorticity plus straight flow can be described as ... 

At lower levels of U strength, the center of the vortex is lower. In fact, the center of the vortex is equivalent to what the U stregth is.  

Therefore at upper levels, the stronger the U component becomes, the center of the vortex is pushed further and further towards the positive y direction. With a high enough U strength, the v component becomes negligable and the u component overrides, thus appearing as just a constant u wind. 

Relating this back to atmospheric patterns, there stronger the U wind is, the further north the center of the votrex is.

## Answer to Exercise 5: 

When you combine divergence and deformation you get ... 

In the case of divergence and deformation 1, you get a very interesting and slightly surprising pattern. With pure divergence and pure deformation, the v components of each neglect each other, and you are left with solely u component wind. There is a "zero line" of no wind at the negative U strength, in this case 0. Away from this line, the magnitude of the wind (u-component) increases exponentially.

In the case of divergence and deformation 2, you get an interesting, yet less surprising pattern. There is a "zero line" at y = -x - U_stregth (y = -x - 0 in this case). Positive of this line (up and right), the u and v components both approach being equal and increasing. Negative of this line (down and left), the u and v components approach equating and becoming more negative. 

In both cases there is clear divergence of the wind, yet only with deformation 2 is the deformation of the wind still easily visual.
